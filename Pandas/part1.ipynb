### Basic Tips for reading files in pandas

```python
# Basic information

pd.read_excel('datasets/world_population_excel_workbook.xlsx').info()
```

```python
# print Shape
pd.read_excel('datasets/world_population_excel_workbook.xlsx').shape
```

```python
# Print first 5 rows
pd.read_excel('datasets/world_population_excel_workbook.xlsx').head()
```

```python
# Print last 5 rows
pd.read_excel('datasets/world_population_excel_workbook.xlsx').tail()
```

```python
# Print random 5 rows
pd.read_excel('datasets/world_population_excel_workbook.xlsx').sample(5)
```

```python
# print specific row
pd.read_excel('datasets/world_population_excel_workbook.xlsx').loc[40]
```
